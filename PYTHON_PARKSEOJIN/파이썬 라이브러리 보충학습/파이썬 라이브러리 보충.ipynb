# train, test Data 불러오기
   - passengerid 컬럼을 index

```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv', index_col = 'PassengerId')
test = pd.read_csv('test.csv', index_col = 'PassengerId')
```

```python
# 데이터 확인
# info()
train.info()
```

```python
# 각 컬럼에 결측치가 몇개있나 합계 내보기
# null인가 아닌가 비교해주는 함수 : isnull()
# 합계를 내주는 함수 : sum()

train.isnull() # 결측치를 가지고 있으면 True가 반환된다.
```

```python
train.isnull().sum() # sum함수 그냥 썼을때 열들의 합을 나타내준다.
```

```python
test.isnull().sum()
```

```python
# Survived 1 : 살았다 0 : 죽었다
# 산 사람들의 데이터만 출력
train[train["Survived"] == 1]
```

```python
# 산 사람들의 데이터 중에서 성별 컬럼 불러와서 항목의 갯수 세기
Survived_1 = train[train["Survived"] == 1]['Sex'].value_counts()
```

```python
Survived_0 = train[train["Survived"] == 0]['Sex'].value_counts()
```

```python
# Survived_1, Survived_0 을 이용해 DataFrame 만들기
# index = Survived, Dead
# columns = Female, male
# 컬럼명주면서 데이터 프레임 만들기
pd.DataFrame({"Survived" : Survived_1, "Dead" : Survived_0})
```

```python
# 행과 열을 바꿔준 데이터를 result안에 넣기
result = pd.DataFrame({"Survived" : Survived_1, "Dead" : Survived_0}).T
```

```python
result
```

```python
# 선생님 방식
# 데이터 프레임에 들어갈 자료들 나열
pd.DataFrame([Survived_1, Survived_0])
```

```python
# 인덱스명을 지정해 주면서 데이터 프레임 만들기
result = pd.DataFrame([Survived_1, Survived_0], index = ['Survived', 'Dead'])
```

```python
result
```

```python
# rename : 컬럼명 바꾸는 법
result.rename(columns = {'female' : '여자', 'male' : '남자'})
```

```python
# 값 변환하기 : mapping(매핑)
# 성별컬럼의 값을
# 여자는 1
# 남자는 0
```

```python
# 성별컬럼의 값이 남자인 데이터 불러오기
# 성별컬럼의 값이 여자인 데이터 불러오기
# 그 데이터들에 값을 입력해줌
# loc[행,열] : 정확한 행과 열의 이름으로 쓸때
train.loc[train['Sex'] == 'female', 'Sex'] = 1
train.loc[train['Sex'] == 'male', 'Sex'] = 0
```

```python
train
```

```python
# Embarked중 결측치가 존재하는 행 확인하기, 한번 더 결측치를 가진 행이 true로 반환된다. 채워져있으면 false
```

```python
train[train["Embarked"].isnull()]
```

```python
# Embarked 특성의 데이터 갯수 세보기
train["Embarked"].value_counts()
```

```python
# Embarked특성의 결측치 부분을 'S'로 채우기
# 내 방식 : 해당 부분을 매핑해서 채워넣음
train[train["Embarked"].isnull()] = 'S'
```

```python
train["Embarked"].value_counts()
```

```python
# 선생님 방식 : fillna이용, #inplace 쓰지 않으면 채워서 한번 보여주고 끝이지 수정되지는 않는다.
train['Embarked'].fillna('S', inplace = True)
train['Embarked'] = train['Embarked'].fillna('S')
```

```python
train.info()
```

```python
# SibSp : 형제자매수
# Parch : 부모자식수
# FamilySize 특성 만들기
# FamilySize = SibSp + Parch + 1
```

```python
train = pd.read_csv('train.csv', index_col = 'PassengerId')
```

```python
FamilySize = train['SibSp'] + train['Parch'] + 1
```

```python
del train['FamilySize']
```

```python
# FamilySize 변수를 DataFrame으로 만들고 
# Family와 train을 concat함수를 통해서 합치기
```

```python
FamilySize = pd.DataFrame(FamilySize)
```

```python
FamilySize
```

```python
FamilySize.rename(columns={0:"FamilySize"}, inplace = True)
```

```python
pd.concat([FamilySize, train], axis = 1)
```
