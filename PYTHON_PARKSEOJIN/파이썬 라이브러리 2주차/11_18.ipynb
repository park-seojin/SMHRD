```python
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0', 'B1', 'B2', 'B3'],
                    'C' : ['C0', 'C1', 'C2', 'C3'],
                    'D' : ['D0', 'D1', 'D2', 'D3']}, 
                   index = [0, 1, 2, 3])

df2 = pd.DataFrame({'A' : ['A4', 'A5', 'A6', 'A7'],
                    'B' : ['B4', 'B5', 'B6', 'B7'],
                    'C' : ['C4', 'C5', 'C6', 'C7'],
                    'D' : ['D4', 'D5', 'D6', 'D7']}, 
                   index = [0, 1, 2, 3])

df3 = pd.DataFrame({'A' : ['A8', 'A9', 'A10', 'A11'],
                    'B' : ['B8', 'B9', 'B10', 'B11'],
                    'C' : ['C8', 'C9', 'C10', 'C11'],
                    'D' : ['D8', 'D9', 'D10', 'D11']}, 
                   index = [0, 1, 2, 3])
```

```python
# 인덱스 번호 같게 해주면 제대로 병합된다.
pd.concat([df1, df2, df3], axis = 1)
```

```python
# 다시 원래대로
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0', 'B1', 'B2', 'B3'],
                    'C' : ['C0', 'C1', 'C2', 'C3'],
                    'D' : ['D0', 'D1', 'D2', 'D3']}, 
                   index = [0, 1, 2, 3])

df2 = pd.DataFrame({'A' : ['A4', 'A5', 'A6', 'A7'],
                    'B' : ['B4', 'B5', 'B6', 'B7'],
                    'C' : ['C4', 'C5', 'C6', 'C7'],
                    'D' : ['D4', 'D5', 'D6', 'D7']}, 
                   index = [4, 5, 6, 7])

df3 = pd.DataFrame({'A' : ['A8', 'A9', 'A10', 'A11'],
                    'B' : ['B8', 'B9', 'B10', 'B11'],
                    'C' : ['C8', 'C9', 'C10', 'C11'],
                    'D' : ['D8', 'D9', 'D10', 'D11']}, 
                   index = [8, 9, 10, 11])
```

```python
df4 = pd.DataFrame({'B' : ['B2', 'B3', 'B6', 'B7'],
                    'D' : ['D2', 'D3', 'D6', 'D7'],
                    'F' : ['F2', 'F3', 'F6', 'F7']},
                    index = [2, 3, 6, 7])
```

```python
# outer join
# 모든 것을 합치겠다.
result = pd.concat([df1, df4], axis = 1)
result
```

```python
# inner join
# 병합할 데이터의 공통된 인덱스 값만 추출
result = pd.concat([df1, df4], axis = 1, join = 'inner')
result
```

```python
# 파이썬 버젼과 pandas 버젼이 다를때 오류 생기기도 한다. 이럴 땐 밑에 처럼 하기
result = pd.concat([df1, df4], axis = 1, join_axes = [df1.index])
result
```

```python
# df1과 df4의 아웃터 조인한 값 중에서 df1에 해당하는 부분만 가져옴
result = pd.concat([df1, df4], axis = 1).reindex(df1.index)
result
```

```python
# df1과 df4의 아웃터 조인한 값 중에서 df4에 해당하는 부분만 가져옴
result = pd.concat([df1, df4], axis = 1).reindex(df4.index)
result
```

```python
# ignore_index
# index값들을 무시하고 데이터프레임을 병합
result = pd.concat([df1, df4], ignore_index = True)
result
```

```python
# merge 함수

df5 = pd.DataFrame({'key' : ['K0', 'K2', 'K3', 'K4'],
                    'A' : ['A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0', 'B1', 'B2', 'B3']})

df6 = pd.DataFrame({'key' : ['K0', 'K1', 'K2', 'K3'],
                    'C' : ['C0', 'C1', 'C2', 'C3'],
                    'D' : ['D0', 'D1', 'D2', 'D3']})
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5, df6, on = "key")
```

```python
# how속성
# left, right, outer, inner
pd.merge(df5, df6, how = 'left', on = 'key')
```

```python
pd.merge(df5, df6, how = 'right', on = 'key')
```

```python
# merge에서는 key값의 모든 것을 기준으로 나열한다.
pd.merge(df5, df6, how = 'outer', on = 'key')
```

```python
pd.merge(df5, df6, how = 'inner', on = 'key')
```

```python
# 1. 데이터 프레임 형성 : pd.DataFrame
# 데이터 프레임 함수에서는 axis 사용 못한다.
# 데이터 프레임 함수에서는 모든 데이터들을 같은 타입으로 처리를 해주기 때문에 실수형으로 출력된다. 이를 타입 바꿔주는 작업 필요
result = pd.DataFrame([s1, s2_s1, s2, s3_s2, s3]).T
```
