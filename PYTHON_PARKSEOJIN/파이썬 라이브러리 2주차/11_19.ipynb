```python
# 기본 plot그래프
# line style > ls
# marker > marker
# line width > lw
# maker size > ms

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y, ls = '--', marker = '*', lw = 5, ms = 20, color = 'silver')
plt.show()
```

```python
# 그림 범위 지정
# xlim : x축 범위 지정
# ylim : y축 범위 지정

x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]
plt.plot(x, y, ls = '--', marker = 'o', mec = 'green', mew = 5, mfc = 'red', lw = 5, ms = 20)

plt.xlim(-2, 10)
plt.ylim(-5, 12)

plt.show()

```

```python
# 틱 설정

x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]
plt.plot(x, y, ls = '--', marker = 'o', mec = 'green', mew = 5, mfc = 'red', lw = 5, ms = 20)

plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])

plt.show()

```

```python
data = pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc_kr')
```

```python
data["요일"].value_counts()
```

```python
temp = data["요일"].value_counts()
```

```python
x = ["월", "화", "수", "목", "금", "토", "일"]
y = temp[["월", '화', "수", "목", "금", "토", "일"]]
```

```python
# bar 차트
# plt.bar
plt.bar(x, y)
plt.show()
```

```python
# x축에 요일
# y축에 사고건수
# 제목 요일별 교통사고 건수

plt.title('요일별 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.bar(x, y)
plt.plot(x, y)
plt.xticks([1, 2, 3, 4, 5, 6, 7], ["월", "화", "수", "목", "금", "토", "일"])
#plt.xticks(x, ["월", "화", "수", "목", "금", "토", "일"])


plt.figure(figsize = (15, 8))
plt.plot(x, y)
plt.xlabel("지역")
plt.ylabel("사상자수")
plt.title("차 VS 차 교통사고의 사상자 수")
plt.show()
```

```python
#차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화 해보기
# 1. 사고유형_대분류 칼럼에서 차대차 사건만 인덱싱
# 2. 1번 데이터에서 사상자 수와 발생지 시도 데이터를 인덱싱
# 3. 2번 데이터에서 발생지시도로 그룹화하고 그룹화된 데이터의 합계를 구함
# 4. plot그래프 그리기

#1번 답
car_to_car = data[data["사고유형_대분류"] == "차대차"]
```

```python
#2번 답
car_to_car_row = car_to_car[["사상자수" , "발생지시도"]]
```

```python
car_to_car_row 
```

```python
# 3번힌트
#groupby()
#sum()
# 3번 답
car_to_car_row.groupby("발생지시도").sum()
```

```python
car_to_car_group = car_to_car_row.groupby("발생지시도").sum()
```

```python
x = car_to_car_group.index
y = car_to_car_group.values
```

```python
plt.figure(figsize = (15, 8))
plt.plot(x, y)
plt.xlabel("지역")
plt.ylabel("사상자수")
plt.title("차 VS 차 교통사고의 사상자 수")
plt.show()
```
